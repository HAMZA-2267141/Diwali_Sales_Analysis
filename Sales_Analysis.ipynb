```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# Exploratory Data Analysis
```

```python
#Gender
```

```python
Gender_EDA = sns.countplot(x='Gender' , data = df)
```

```python
#axes : xlabel = 'Gender' , yLabel = 'count'>
```

```python
Gender_EDA = sns.countplot(x='Gender' , data = df)
for bars in Gender_EDA.containers:
    Gender_EDA.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
Sales_Gender = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender' , y='Amount' , data = Sales_Gender)
```

```python
sns.countplot(x='Age Group' , data = df )
```

```python
AgeGroup_EDA = sns.countplot(x='Age Group' , data = df , hue = 'Gender')
```

```python
AgeGroup_EDA = sns.countplot(x='Age Group' , data = df , hue = 'Gender')
for bars in AgeGroup_EDA.containers:
    AgeGroup_EDA.bar_label(bars)
```

```python
Sales_Age = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group' , y='Amount' , data = Sales_Age)
```

```python
#State
```

```python
df.groupby(['State'],  as_index = False  )['Orders'].sum().sort_values(by='Orders' , ascending = False)
```

```python
Sales_State = df.groupby(['State'],  as_index = False  )['Orders'].sum().sort_values(by='Orders' , ascending = False).head(8)
sns.set(rc={'figure.figsize': (15,5)})
sns.barplot(x = 'State' , y = 'Orders' , data = Sales_State)
```

```python
df.groupby(['State'],  as_index = False  )['Amount'].sum().sort_values(by='Amount' , ascending = False)
```

```python
Sales_StateAmont = df.groupby(['State'],  as_index = False  )['Amount'].sum().sort_values(by='Amount' , ascending = False).head(8)
sns.set(rc= {'figure.figsize' : (15,5)})
sns.barplot(x = 'State' , y  = 'Amount' , data = Sales_StateAmont)
```

```python
Marital_status_EDA = sns.countplot(x = 'Marital_Status' , data = df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in Marital_status_EDA.containers:
    Marital_status_EDA.bar_label(bars)
```

```python
df.groupby(['Marital_Status' , 'Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
```

```python
Sales_MaritalStatus_Amount = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Marital_Status' , y= 'Amount' , hue = 'Gender' , data  =Sales_MaritalStatus_Amount)
```

```python
df.groupby(['Marital_Status' , 'Gender'], as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False)
```

```python
Sales_MaritalStatus_Orders = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False)
sns.barplot(x = 'Marital_Status' , y= 'Orders' , hue = 'Gender' , data  =Sales_MaritalStatus_Orders)
```

```python
sns.countplot(x='Occupation' , data =df)
sns.set(rc={'figure.figsize':(20,5)})
```

```python
Occupation_EDA = sns.countplot(x='Occupation' , data =df)
sns.set(rc={'figure.figsize':(20,5)})
for bars in Occupation_EDA.containers:
    Occupation_EDA.bar_label(bars)
```

```python
df.groupby(['Occupation', 'Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
```

```python
sales_OccupationAmount = df.groupby(['Occupation', 'Gender'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending = False)
sns.set(rc={'figure.figsize' :(22,5)})
sns.barplot(x = 'Occupation' , y = 'Amount' , data = sales_OccupationAmount , hue = 'Gender')
```

```python
sales_OccupationOrders = df.groupby(['Occupation', 'Gender'] , as_index = False)['Orders'].sum().sort_values(by = 'Orders' , ascending = False)
sns.set(rc={'figure.figsize' :(22,5)})
sns.barplot(x = 'Occupation' , y = 'Orders' , data = sales_OccupationOrders , hue = 'Gender')
```

```python
Sales_Occupation = df.groupby(['Occupation'] , as_index = False)['Amount'].sum().sort_values(by = 'Amount' , ascending  = False)
sns.set(rc = {'figure.figsize' : (22 , 5)})
sns.barplot( x = 'Occupation' , y = 'Amount' , data = Sales_Occupation)
```
